Смотрим типы данных, температуру и pH необходимо перезаписать в числовой тип (float).

```python
def process_intervals_fixed(element):
  if isinstance(element, str) and '-' in element:
    splitted = element.split('-')
    average = (float(splitted[0]) + float(splitted[1])) / 2

    return average

  else:
    return pd.to_numeric(element, errors='coerce')

df3['pH'] = df3['pH'].apply(process_intervals_fixed)
```

Колонки содержат 2 параметра, разделенных '-', необходимо заменить на среднее значение. В колонке температуры повторяем то же самое и удаляем значок градуса.

```python
def process_temperature(element):
  if not isinstance(element, str):
    return element

  cleaned_element = element.replace('°C', '').strip()

  if '-' in cleaned_element:
    try:
      splitted = cleaned_element.split('-')
      average = (float(splitted[0]) + float(splitted[1])) / 2
      return average

    except (ValueError, IndexError):
      return np.nan

  else:
    return pd.to_numeric(cleaned_element, errors='coerce')

df3['Temperature'] = df3['Temperature'].apply(process_temperature)
```

```python
obj_columns = ['pH', 'Temperature']

df3['Temperature'] = df3['Temperature'].replace('-----', np.nan)
df3['pH'] = df3['pH'].replace('-----', np.nan)

for i in obj_columns:
    df3[i] = df3[i].astype(float)
```

Преобразовали колонки в тип float.

```python
Q1 = df3['pKm_value'].quantile(0.25)
Q3 = df3['pKm_value'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

current_outliers_indices = df3[(df3['pKm_value'] < lower_bound) | (df3['pKm_value'] > upper_bound)].index

df3['pKm_value'] = np.where(
    df3['pKm_value'] > upper_bound,
    upper_bound,
    df3['pKm_value']
)
df3['pKm_value'] = np.where(
    df3['pKm_value'] < lower_bound,
    lower_bound,
    df3['pKm_value']
)

print("Значения после замены:")
print(df3.loc[current_outliers_indices, ['pKm_value']])
```

```python
desc_col = ['RingCount', 'AromaticRings', 'HeteroAtoms', 'AliphaticRings', 'SaturatedRings', 'RotatableBonds',
                  'AmideBonds', 'HBD', 'HBA', 'NHOHCount', 'NumHeterocycles', 'MolMR', 'TPSA', 'PEOE_VSA1', 'PEOE_VSA2', 'PEOE_VSA3', 'PEOE_VSA4',
            'PEOE_VSA5', 'PEOE_VSA6', 'PEOE_VSA7', 'PEOE_VSA8',
             'PEOE_VSA9', 'PEOE_VSA10', 'PEOE_VSA11', 'PEOE_VSA12', 'PEOE_VSA13', 'SMR_VSA1', 'SMR_VSA2', 'SMR_VSA3',
             'SMR_VSA4', 'SMR_VSA5', 'num_chains', 'num_residues', 'molecular_weight', 'isoelectric_point', 'instability_index', 'gravy', 'radius_of_gyration',
          'num_disulfide_bonds', 'AACount_A', 'AACount_C', 'AACount_D', 'AACount_E', 'AACount_F', 'AACount_G', 'AACount_H', 'AACount_I',
           'AACount_K', 'AACount_L', 'AACount_M', 'AACount_N', 'AACount_P', 'AACount_Q', 'AACount_R',
           'AACount_S', 'AACount_T', 'AACount_V', 'AACount_W', 'AACount_Y']
for i in desc_col:
  Q1 = df3[i].quantile(0.25)
  Q3 = df3[i].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  current_outliers_indices = df3[(df3[i] < lower_bound) | (df3[i] > upper_bound)].index

  df3[i] = np.where(
      df3[i] > upper_bound,
      upper_bound,
      df3[i]
  )
  df3[i] = np.where(
      df3[i] < lower_bound,
      lower_bound,
      df3[i]
  )
```

Провели стандартизацию числовых значений.

```python
km = ['Km_value (mM)', 'pKm_value']
scaler = StandardScaler()
df3[km] = scaler.fit_transform(df3[km])
```

```python
categorical_features = ['EC_number', 'Organism_name', 'Site_type']

df4 = pd.get_dummies(df3, columns=categorical_features, prefix=categorical_features, drop_first=True)
```

```python
target_column = 'pKm_value'
y = df4[target_column]

X = df4.select_dtypes(include=np.number)


if target_column in X.columns:
    X = X.drop(columns=target_column)

corr_matrix = X.corr().abs()
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.90)]

print(f"Найдено {len(to_drop)} сильно коррелирующих признаков для удаления:")
print(to_drop)

X_reduced = X.drop(columns=to_drop)

print(f"\nРазмер данных до удаления коррелирующих признаков: {X.shape}")
print(f"Размер данных после удаления (готово для модели): {X_reduced.shape}")


final_df = pd.concat([X_reduced, y], axis=1)
final_df
```

Произвели кодирование категориальных переменных методом one hot coding.

```python
temp_df = pd.concat([X_reduced, y], axis=1)

correlations_with_target = temp_df.corr()['pKm_value'].abs().sort_values(ascending=False)

print("\nКорреляция признаков с pKm_value:")
print(correlations_with_target.head(10))


relevant_features = correlations_with_target[correlations_with_target > 0.05].index.tolist()

if 'pKm_value' in relevant_features:
    relevant_features.remove('pKm_value')

print(f"\nОставляем {len(relevant_features)} признаков с корреляцией > 0.05")


X_final_selection = X_reduced[relevant_features]

print(f"\nРазмер данных до отбора: {X_reduced.shape}")
print(f"Размер данных после отбора: {X_final_selection.shape}")
```
